```python
import pandas as pd
```

```python
df = pd.read_csv("6m_0k_99k_users_dataset_public.csv")
```

```python
df
```

```python
df.head()
```

```python
chunk_size = len(df) // 5 + (1 if len(df) % 5 else 0)
for i in range(5):
    start = i * chunk_size
    end = start + chunk_size
    chunk_df = df.iloc[start:end]
    chunk_df.to_csv(f'chunk-data/chunk{i+1}.csv', index=False)
chunk_df
```
